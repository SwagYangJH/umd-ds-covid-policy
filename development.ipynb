## 3.1 Cleaning up Mobility Data 

```python
df_mobility_usa = df_mobility[df_mobility['country']=='United States']
```

```python
# Melting in order to transpose 
df_mobility_usa_melted = df_mobility_usa.melt(id_vars=["region", "transportation_type","sub-region","country","geo_type"], 
        var_name="Date", 
        value_name="Value")
```

```python
DF_MOBILITY_MAIN = df_mobility_usa_melted[df_mobility_usa_melted['Date']!='alternative_name']
```
